### imports and functions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import csv

def readcsv(filename):
    Points = []
    p = []
    with open(filename) as points:
        read = csv.reader(points, delimiter=',')   
        for row in read:
            P = row[1]
            p.append(P)
    for i in p:
        Points.append(float(i))
    Points.pop(0)
    return Points
```

### plot logarithmic

```python
pureXlog = readcsv('nonlinearSimsDataXpureLogarithmic.csv')
pureYlog = readcsv('nonlinearSimsDataypureLogarithmic.csv')
noiseXlog = readcsv('nonlinearSimsDataXnoiseLogarithmic.csv')
noiseYlog = readcsv('nonlinearSimsDataynoiseLogarithmic.csv')

sns.scatterplot(pureXlog,pureYlog,color = 'k')
sns.scatterplot(noiseXlog,noiseYlog,color = 'r')
```

### plot sine period 4pi

```python
pureXsine = readcsv('nonlinearSimsDataXpureSine Period $4\pi$.csv')
pureYsine = readcsv('nonlinearSimsDataypureSine Period $4\pi$.csv')
noiseXsine = readcsv('nonlinearSimsDataXnoiseSine Period $4\pi$.csv')
noiseYsine = readcsv('nonlinearSimsDataynoiseSine Period $4\pi$.csv')

sns.scatterplot(pureXsine,pureYsine,color = 'k')
sns.scatterplot(noiseXsine,noiseYsine,color = 'r')
```

### plot square

```python
pureXsq = readcsv('nonlinearSimsDataXpureSquare.csv')
pureYsq = readcsv('nonlinearSimsDataypureSquare.csv')
noiseXsq = readcsv('nonlinearSimsDataXnoiseSquare.csv')
noiseYsq = readcsv('nonlinearSimsDataynoiseSquare.csv')

sns.scatterplot(pureXsq,pureYsq,color = 'k')
sns.scatterplot(noiseXsq,noiseYsq,color = 'r')
```

### plot multiplicative

```python
pureXmult = readcsv('nonlinearSimsDataXpureMultiplicative.csv')
pureYmult = readcsv('nonlinearSimsDataypureMultiplicative.csv')

sns.scatterplot(pureXmult,pureYmult,color = 'k')
```

### plot independence

```python
pureXinde = readcsv('nonlinearSimsDataXpureIndependence.csv')
pureYinde = readcsv('nonlinearSimsDataypureIndependence.csv')

sns.scatterplot(pureXinde,pureYinde,color = 'k')
```

### sublplotted

```python
plt.figure(figsize = (20, 4))
ax = plt.subplot(1,5,1)
sns.scatterplot(pureXlog,pureYlog,color = 'k',linewidth = 0,ax = ax)
sns.scatterplot(noiseXlog,noiseYlog,color = 'r',linewidth = 0,ax = ax)
ax.title.set_text('Logarithmic')
ax.set_frame_on(False) #Remove both axes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax = plt.subplot(1,5,2)
sns.scatterplot(pureXsine,pureYsine,color = 'k',linewidth = 0,ax = ax)
sns.scatterplot(noiseXsine,noiseYsine,color = 'r',linewidth = 0,ax = ax)
ax.title.set_text('Sine $4\pi$')
ax.set_frame_on(False) #Remove both axes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax = plt.subplot(1,5,3)
sns.scatterplot(pureXsq,pureYsq,color = 'k',linewidth = 0,ax = ax)
sns.scatterplot(noiseXsq,noiseYsq,color = 'r',linewidth = 0,ax = ax)
ax.title.set_text('Square')
ax.set_frame_on(False) #Remove both axes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax = plt.subplot(1,5,4)
sns.scatterplot(pureXmult,pureYmult,color = 'k',linewidth = 0,ax = ax)
ax.title.set_text('Multiplicative')
ax.set_frame_on(False) #Remove both axes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax = plt.subplot(1,5,5)
sns.scatterplot(pureXinde,pureYinde,color = 'k',linewidth = 0,ax = ax)
ax.title.set_text('Independence')
ax.set_frame_on(False) #Remove both axes
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)

ax.figure.savefig("plotted_simulations.png")
```
