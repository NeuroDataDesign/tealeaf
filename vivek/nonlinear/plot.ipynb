```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_context("talk")
```

```python
df = pd.read_csv("results.csv", index_col="Unnamed: 0")
df.head()
```

```python
sns.relplot(
    x="n_samples",
    y="score",
    hue="criterion",
    col="simulation",
    kind="line",
    data=df,
    facet_kws={'sharey': False, 'sharex': True}
)
plt.show()
```

```python
g = sns.relplot(
    x="n_samples",
    y="runtime",
    hue="criterion",
    col="simulation",
    kind="line",
    data=df,
)

g.set(ylabel="Runtime (s)")

plt.show()
```
