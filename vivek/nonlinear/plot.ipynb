```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_context("talk", font_scale=1.5)
```

```python
df = pd.read_csv("results/sim_1/sim_1.csv", index_col="Unnamed: 0")

sns.relplot(
    x="n_samples",
    y="average",
    hue="criterion",
    col="simulation",
    kind="line",
    data=df,
    facet_kws={'sharey': False, 'sharex': True}
)
plt.savefig('sim_1.png', dpi=400)
```
